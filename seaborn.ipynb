```python
 df1=pd.read_csv('/content/Who_is_responsible_for_global_warming.csv')
```

```python
df1
```

```python
df2=df1.drop(columns=['Country Code','Indicator Name','Indicator Code'],axis=1)
```

```python
df2=df2.set_index('Country Name')
```

```python
sns.heatmap(df2)
```

```python
sns.heatmap(df2,cmap='CMRmap')
```

```python
plt.figure(figsize=(16,9))
sns.heatmap(df2,cmap='CMRmap',annot=True)
```

```python
plt.figure(figsize=(16,9))
annot_kws={
    'fontsize':15,
    'fontstyle':'italic',
    'color':'k',
    'rotation':'vertical',
    'verticalalignment':'center',
    'backgroundcolor':'g'
}
sns.heatmap(df2,cmap='CMRmap',annot=True,annot_kws=annot_kws)
```

```python
plt.figure(figsize=(16,9))
sns.heatmap(df2,cmap='CMRmap',annot=True,linewidth=4,linecolor='w')
```

```python
plt.figure(figsize=(16,9))
cbar_kws={
    'orientation':'horizontal',
    'shrink':1,
    'extend':'min',
    'extendfrac':0.1,
    'ticks':np.arange(0,22),
    'drawedges':True
}
sns.heatmap(df2,cmap='YlOrRd_r',annot=True,linewidth=4,linecolor='w',cbar_kws=cbar_kws)
```

```python
plt.figure(figsize=(16,9))
ax=sns.heatmap(df2.corr(),annot=True,linewidth=3)
#ax.tick_params(size=10,color='w',labelsize=10,labelcolor='w')
```
