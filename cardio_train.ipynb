```python
df=df.drop(['yr'],axis=1)
```

```python
df=df.drop(['id'],axis=1)
```

```python
x=df.iloc[:,:-1]
x
```

```python
y=df.iloc[:,-1]
y
```

```python
from sklearn.ensemble import ExtraTreesClassifier
```

```python
model=ExtraTreesClassifier()
```

```python
model.fit(x,y)
```

```python
model.feature_importances_
```

```python
top=pd.Series(model.feature_importances_,index=x.columns)
top
```

```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import f_classif

```

```python
rg=SelectKBest(score_func=f_classif)
```

```python
rg.fit(x,y)
```

```python
new=pd.DataFrame(rg.scores_,columns=['score value'])
new
```

```python
r=pd.Series(rg.scores_,index=x.columns)
```

```python
new=pd.DataFrame(r,columns=['score value'])
new
```

```python
name=pd.DataFrame(x.columns)
```

```python
new_f=pd.concat([name,new],axis=1)
new_f
```

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain,xtest,ytrain,ytest,=train_test_split(x,y,test_size=.30,random_state=1)
```

```python
xtrain
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rfc=RandomForestClassifier()
```

```python
rfc.fit(xtrain,ytrain)
```

```python
rfc.score(xtest,ytest)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
clf=DecisionTreeClassifier()
```

```python
clf.fit(xtrain,ytrain)
```

```python
clf.score(xtest,ytest)
```
