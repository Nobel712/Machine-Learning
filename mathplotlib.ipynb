```python
class1=['Python','R','Al','ML','DL','Ds']
clas1_student=[30,10,20,25,10,35]
class2_student=[40,4,34,20,10,24]
class3_student=[35,6,23,45,32,21]
```

```python
plt.barh(class1,clas1_student)
```

```python
plt.bar(class1,class2_student,width=0.5,align='edge',color='r',edgecolor='b',linewidth=1.5,alpha=0.9,linestyle=':')
```

```python
plt.figure(figsize=(16,9))
class2=np.arange(len(class1))
width=0.3
plt.bar(class2,clas1_student,color='b',width=0.3,label='clas1_student')
plt.bar(class2+width,class2_student,color='r',width=0.3,label='class2_student')
plt.bar(class2+width+width,class3_student,color='g',width=0.3,label='class3_student')
plt.xticks(class2+width,class1)
plt.legend()
```

```python
classes =['python','R','Al','ML','DL',"DS"]
class1_student=[30,10,20,25,10,23]
```

```python
expode=[0.3,0,0,0,0,0]
plt.pie(class1_student,labels=classes,explode=expode,autopct='%0.2f%%',shadow=True,radius=1.4)
plt.savefig("ddd.png")
plt.show()
```

```python
expode=[0,0,0,0,0.3,0]
color=['r','b','g','y','m','c']
textprops={'fontsize':15}
plt.pie(class1_student,labels=classes,explode=expode,colors=color,radius=1.4,shadow=True,startangle=90,autopct='%0.2f%%',textprops=textprops)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import style
```

```python
style.use('ggplot')
classes=['python','R','Al','ML','DS']
class1=[30,10,20,25,10]
class2=[40,5,20,20,10]
class3=[35,5,30,15,15]
plt.bar(classes,class1)
```

```python
style.use('ggplot')
plt.bar(classes,class1, align='edge',width=.5,color='r',edgecolor='b',linestyle=':',linewidth=1.5,alpha=0.5)
```

```python
plt.figure(figsize=(16,9))
classes1=np.arange(len(classes))
width=0.2
plt.bar(classes1,class1,width,linewidth=1.5,edgecolor='b',linestyle=':')
plt.bar(classes1+width,class2,width,linewidth=1.5,edgecolor='b',linestyle=':')
plt.bar(classes1+(2*width),class3,width,linewidth=1.5,edgecolor='b',linestyle=':')
plt.xticks(classes1+width,classes,rotation=20)
plt.show()
```
